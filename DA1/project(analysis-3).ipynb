```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random as rd
import sys
import squarify as sq
import unicodedata
```

```python
data=pd.read_csv(r"C:\Users\KATYAYANI\Downloads\autos.csv",encoding_errors="replace")
```

```python
data1=data.copy()
```

```python
data1=data1.drop_duplicates()
```

```python
data1 = data1.fillna("Not mentioned")
```

```python
data1
```

```python
#2.What is the Average price of vehicle by fuel type and gearbox type.Give a plot

k = pd.pivot_table(data1, index ="fuelType", columns="gearbox", values="price", aggfunc=["mean"])
k
```

```python
sns.heatmap(k, annot=True)
plt.title("Average price of vehicle by fuel type and gearbox type\n")
plt.show()
```

```python
#3.What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot
k= pd.pivot_table(data1, index="vehicleType", columns="gearbox", values="powerPS", aggfunc="mean")
k
```

```python
sns.heatmap(k, annot=True)
plt.title("Average power of a vehicle by vehicle type and gearbox type\n")
plt.show()
```

```python
#4.What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this
k = pd.pivot_table(data1, index="brand", columns="vehicleType", values="price", aggfunc="mean")
k
```

```python
sns.heatmap(k)
plt.title("Average price of a vehicle by brand as well as vehicle type\n")
plt.show()
```
